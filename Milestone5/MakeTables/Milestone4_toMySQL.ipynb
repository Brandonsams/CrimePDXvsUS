```python
# create the base url
url = f'https://api.usa.gov/crime/fbi/sapi/api/estimates/national/1960/2020?API_KEY={accessKey}'
```

```python
import requests

# Query the api
r = requests.get(url)
j = r.json()

# Make the json valid
j_str = str(j).replace("'",'"').replace("None",'"None"')
```

```python
# write the json to a file
with open('crime_data.json', 'w') as f:
    f.write(j_str)
```

```python
# convert the json into a dataframe

import pandas as pd
import json
data = json.load(open('crime_data.json'))
df = pd.DataFrame(data["results"])

df
```

```python
# Make the year an index, and sort by year
df = df.set_index('year').sort_index()

df
```

```python
# Replace the None string with NaN, for missing values
for col in df:
    df[col].replace('None', pd.np.nan, inplace=True)

df
```

```python
# Remove columns that have to do with state, as these are national statistics
df = df.drop(columns = ['state_id','state_abbr'])

df
```

```python
# Convert all columns to Int64
# By default, the numbers were int64, not Int64. This change allows integer-columns to have null data, <NA>.
df = df.astype('Int64')

df
```

```python
# create a per-capita column for every measure of crime
for col in df:
    if not col == 'population':
        df[f'{col}_per_capita'] = df[col].div(df.population)

df
```

```python
df.index = df.index.astype(int)
```
